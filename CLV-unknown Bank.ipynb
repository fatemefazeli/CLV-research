## step2) creating number of transaction column
In the given dataset of digikala, the order_id is repeat per product_id of same order so count the number of ordr_id row dosn't show the real number of transaction lonly.
In this part we group two columns 'int_user_id' & 'order_id' and get the number of simillar order_id per user_id. So we use .group().size to cuont them and create new DataFrame naming 'sub_df_transaction' at the same time. Then we create new column with  number of transaction title through devided size column to itself. This will give the real number of every transaction meaning '1'.

```python
sub_df_transaction= df_copy.groupby(['int_user_id','order_id']).size().to_frame(name = 'size').reset_index()
sub_df_transaction
```

```python
sub_df_transaction['number of transaction']=sub_df_transaction['size']/sub_df_transaction['size']
sub_df_transaction
```

```python
sub_df_transaction= sub_df_transaction.groupby(['int_user_id']).sum()
sub_df_transaction
```

```python
sub_df_transaction=sub_df_transaction.drop(['order_id','size'],axis=1)
sub_df_transaction
```

## step3) creating quantity of buy column

```python
sub_df_Qobuy= df_copy.groupby(['int_user_id','product_id']).size().to_frame(name = 'quantity of buy').reset_index()
sub_df_Qobuy
```

```python
sub_df_Qobuy=sub_df_Qobuy.groupby(['int_user_id']).sum()
sub_df_Qobuy=sub_df_Qobuy.drop(['product_id'],axis=1)
sub_df_Qobuy
```

## step4) creating total cost column

```python
sub_df_totalcost= df_copy[['int_user_id','price']]
sub_df_totalcost
```

```python
sub_df_totalcost=sub_df_totalcost.groupby(['int_user_id']).sum()
sub_df_totalcost=sub_df_totalcost.rename(columns={'price' : 'total price'})
sub_df_totalcost
```

## step5) creating city & state columns

```python
sub_df_citystate=df_copy.groupby(['int_user_id']).max()
sub_df_citystate=sub_df_citystate.drop(['order_id','variant_id','product_id','quantity','price','date'] , axis=1)
sub_df_citystate
```

## step6) creating target lable

```python
sub_df_frequency= df_copy.groupby(['int_user_id','date']).size().to_frame(name = 'size').reset_index()
sub_df_frequency
```

```python
sub_df_frequency['number of frequency']=sub_df_frequency['size']/sub_df_frequency['size']
sub_df_frequency
```

```python
sub_df_frequency= sub_df_frequency.groupby(['int_user_id']).sum()
sub_df_frequency
```

```python
sub_df_frequency

lable=[]
for i in sub_df_frequency['number of frequency']:
    if i>=2:
        lable.append(1)
    else:
        lable.append(0)
sub_df_frequency['lable']=lable

sub_df_frequency
```

```python
sub_df_frequency=sub_df_frequency.drop(['size','number of frequency'], axis=1)
sub_df_frequency
```

## step7) combine columns to create main data base for CLV

```python
df_main= pd.concat([sub_df_transaction , sub_df_Qobuy , sub_df_totalcost , sub_df_citystate , sub_df_frequency] , axis=1).reindex(sub_df_transaction.index)
df_main
```

```python
df_main.city_id.value_counts()
```

```python
df_main=df_main.drop(['city_id'] , axis=1)
df_main
```

```python
df_main['total price'].max()
Ntotal_price=[]
for i in df_main['total price']:
    i= round((i/df_main['total price'].max())*1000)
    Ntotal_price.append(i)
    
df_main['total price']=Ntotal_price
df_main
    
```
