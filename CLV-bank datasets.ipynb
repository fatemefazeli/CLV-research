### transforming surname to numerical column

```python
df_copy.Surname.value_counts()
```

```python
df_copy['int_Surname']= df_copy['Surname'].astype('category')
df_copy.info()
```

```python
df_copy['int_Surname']= df_copy['int_Surname'].cat.codes
df_copy
```

```python
df_copy=df_copy.drop(['Surname'],axis=1)
df_copy
```

### transforming Geography to numerical column

```python
df_copy.Geography.value_counts()
```

```python
int_Geography=[]
for i in df_copy['Geography']:
    if i=="France":
        int_Geography.append(1)
    elif i=="Germany":
        int_Geography.append(2)
    else :
        int_Geography.append(3)
df_copy['int_Geography']=int_Geography
df_copy=df_copy.drop(['Geography'],axis=1)

df_copy
```

### transforming Gender to numerical column

```python
int_Gender=[]
for i in df_copy['Gender']:
    if i=="Female":
        int_Gender.append(1)
    elif i=="Male":
        int_Gender.append(2)

df_copy['int_Gender']=int_Gender
df_copy=df_copy.drop(['Gender'],axis=1)

df_copy
```

### final preprcessing 

```python
df_copy.info()
```

```python
df_copy.CustomerId.value_counts()
```

```python
df_copy=df_copy.drop(['RowNumber','CustomerId'],axis=1)
```

### Normalizing columns

### Balance

```python
df_copy['Balance'].max()
NBalance=[]
for i in df_copy['Balance']:
    i= round((i/df_copy['Balance'].max())*100)
    NBalance.append(i)
    
df_copy['Balance']=NBalance
df_copy
    
```

### CreditScore

```python
df_copy['CreditScore'].max()
NCreditScore=[]
for i in df_copy['CreditScore']:
    i= round((i/df_copy['CreditScore'].max())*100)
    NCreditScore.append(i)
    
df_copy['CreditScore']=NCreditScore
df_copy
    
```

### EstimatedSalary

```python
df_copy['EstimatedSalary'].max()
NEstimatedSalary=[]
for i in df_copy['EstimatedSalary']:
    i= round((i/df_copy['EstimatedSalary'].max())*100)
    NEstimatedSalary.append(i)
    
df_copy['EstimatedSalary']=NEstimatedSalary
df_copy
```

### int_Surname

```python
df_copy['int_Surname'].max()
Nint_Surname=[]
for i in df_copy['int_Surname']:
    i= round((i/df_copy['int_Surname'].max())*100)
    Nint_Surname.append(i)
    
df_copy['int_Surname']=Nint_Surname
df_copy
```
